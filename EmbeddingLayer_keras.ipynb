```
import keras
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```
#Defining Network with Embedding Layer
embedding_model = keras.Sequential()
embedding_model.add(keras.layers.Embedding(40, 5, input_length=10))
```

```
print(embedding_model.get_weights())
```

```
print(embedding_model.summary())
```

```
#Define input Samples for Embedding Layer
samples = ['The cat sat on the mat.',
           'The dog ate my homework.',
           'My friend name is Tony Singh',
           'I had visited Dubai last year.',
           'I will visit Singapore next year',
           'I am a Software Engineer ',
           'I got good appraisal this year.',
           'I will change Job next year']
```

```
#creates a tokenizer, configured to only take into account the n most common words
tk = Tokenizer(num_words=40, oov_token='OOV')
```

```
#Builds the word index
tk.fit_on_texts(samples)

#Recover the word index that was computed above
word_index = tk.word_index
print('Found %s unique tokens.' % len(word_index))
print(word_index)
```

```
#Turns strings into lists of integer indices
sequences = tk.texts_to_sequences(samples)
print(sequences)
```

```
#Padding
data = pad_sequences(sequences, maxlen=10, padding='post')
print(data)
```

```
#Predicting output
embedding_model.compile(optimizer='rmsprop', loss='mean_absolute_error', metrics=['mean_absolute_error'])
output_array = embedding_model.predict(data)
print(output_array)
print(output_array.shape)
```
